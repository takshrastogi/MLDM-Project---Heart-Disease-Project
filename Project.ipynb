## Data-Preprocessing

```python
# check the presence of missing value

data1.isnull().sum()
```

```python
# handling the missing values from categorical columns()

mode1=  data1['chest_pain_type'].mode()
mode2=  data1['fbs'].mode()
mode3=  data1['thal'].mode()

print('mode of chest_pain_type',mode1)
print('mode of fbs',mode2)
print('mode of thal', mode3)
```

```python
data1['chest_pain_type'].replace(np.nan, ' asymptomatic', inplace=True)
data1['fbs'].replace(np.nan, 'False', inplace=True)
data1['thal'].replace(np.nan, 'normal', inplace=True)
```

```python
# handling the missing values from numerical column()

data1['number_of_major_vessels'].fillna(method= 'ffill', axis=0, inplace=True)
```

```python
data1.isnull().sum()
```

```python
# # extraction of indep and dep variable
x= data1.iloc[:,0:-1]
y= data1.iloc[:,13:14]
```

```python
x
```

```python
y
```

```python
# handling categorical data

X= pd.get_dummies(x)
X.head()
```

```python
y.replace('No disease',0, inplace= True)
y.replace('Disease',1, inplace= True)
```

```python
y
```

```python
# split the dataset

from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test= train_test_split(X,y,test_size=0.2, random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# standardization:
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```
